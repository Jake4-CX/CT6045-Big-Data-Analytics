```python
import pandas as pd
import numpy as np
```

```python
# Read the dataset
dataset = pd.read_csv('../datasets/Dataset-QOE_prediction_ICC2018.csv')
```

```python
# Finding the correlation between 'Stall Events' and other columns
correlation = dataset.corr()['Stall Events']

# Sort correlation in descending order
sorted_correlation = correlation.abs().sort_values(ascending=False)

sorted_correlation.head(10) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Calculate correlation matrix
corr_matrix = dataset.corr()

# Select top 15 features with highest counts of high correlations
threshold = 0.8
high_corr = (corr_matrix.abs() >= threshold).sum().sort_values(ascending=False)
top_15_features = high_corr.head(15).index

# Create correlation matrix for top 15 features
top_corr_matrix = dataset[top_15_features].corr()

# Generate heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(top_corr_matrix, annot=True, fmt=".2f", vmax=1.0, vmin=-1.0, center=0, linewidths=0.5, square=True)
plt.title('Correlation Matrix Heatmap for Top 15 Features')
plt.show()
```
